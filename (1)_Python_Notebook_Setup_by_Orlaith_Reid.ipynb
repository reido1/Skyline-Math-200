```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

Codebook 1: Screenshot 2023-04-27 11.59.36 AM.png

```python
#Manage data for variable 1
```

```python
df['Relationship_Years'] = df['H4RD2Y']
df['Relationship_Years']= pd.to_numeric(df['Relationship_Years'],errors= 'coerce')
df['Relationship_Years']= df['Relationship_Years'].loc[df['Relationship_Years'] != ' ']
df['Relationship_Years']= df['Relationship_Years'].loc[df['Relationship_Years'] < 96]
df['Relationship_Years'].value_counts()
```

```python
df['Relationship_Months']= df['H4RD2M']
df['Relationship_Months']= pd.to_numeric(df['Relationship_Months'],errors= 'coerce')
df['Relationship_Months']= df['Relationship_Months'].loc[df['Relationship_Months'] < 96]
df['Relationship_Months']= df['Relationship_Months'].loc[df['Relationship_Months'] != ' ']
df['Relationship_Months'].value_counts()
```

```python
df['Relationship_Time'] = df['Relationship_Years'] *12+ df['Relationship_Months']
```

```python
#Graphs and tables or statistics
```

```python
sns.histplot(x='Relationship_Time', data=df)
```

```python
sns.boxplot(x= 'Relationship_Time', data=df)
```
