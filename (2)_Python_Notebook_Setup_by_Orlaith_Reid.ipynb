Research Question #2 : Is there an association between education and how happy you are in your relationship? 


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

Codebook 1:Screenshot 2023-04-27 12.05.21 PM.png

```python
#Manage data for variable 1
```

```python
df['H4RD7E'].value_counts()
```

```python
df['relationship_happiness'] = df['H4RD7E']
```

```python
df['relationship_happiness'] = df['relationship_happiness'].loc[ df['relationship_happiness'] !=' ']
df['relationship_happiness'] = df['relationship_happiness'].loc[ df['relationship_happiness'] !='96']
df['relationship_happiness'] = df['relationship_happiness'].loc[ df['relationship_happiness'] !='98']
df['relationship_happiness'].value_counts()
```

```python
df['relationship_happiness'].replace(['1','2','3','4','5'] , ['strongly agree', 'agree', 'neither', 'strongly disagree','disagree'], inplace = True)
df['relationship_happiness'].value_counts()
```

```python
#Graphs and tables or statistics
```

```python

df['relationship_happiness'].value_counts().plot(kind ='pie')
```

```python

df['relationship_happiness'].value_counts(normalize = True).plot(kind ='bar')
```
